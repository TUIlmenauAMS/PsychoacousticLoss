```python
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio

#Test audio file:
fs, snd =wav.read('./Python-Audio-Coder/fantasy-orchestra.wav')
#Audio signal with uniform quantization and de-quantization
#snd_quant=(np.round(snd/10000))*10000
#wav.write('fantasy-orchestra_quant.wav', rate= fs, data=snd_quant.astype(np.int16))

#Show dimension of the audio file:
print("snd.shape=", snd.shape, "np.max(snd)=", np.max(snd), "np.min(snd)=", np.min(snd))
plt.plot(snd)
plt.title("The original sound")
```

```python
#Audio(np.transpose(snd), rate=fs)
Audio(snd[:,0], rate=fs)
#Audio('fantasy-orchestra_quant.wav')
```

```python
Audio(snd[:,1], rate=fs)
```

```python
import psyacloss
ploss=psyacloss.percloss(snd[:,0], snd[:,1], fs)
print("percetual loss =", ploss)
```

```python
import torch
import psyacloss_torch
ploss=psyacloss_torch.percloss(torch.from_numpy(snd[:,0]).float(), torch.from_numpy(snd[:,1]).float(), fs)
print("percetual loss =", ploss)
```
